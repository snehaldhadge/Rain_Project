```python
def rename_dcon_columns(df):
    df = df.rename(columns={'i_Country Code':'d_WBD_councode'
,'i_GDP_Growth':'d_WBD_GDPg'
,'i_GDPperCapitaPPP':'d_WBD_GDPpc'
,'i_Exports_As_GDP':'d_WBD_eGDP'
,'i_FDI':'d_WBD_FDI'
,'i_ServicesValueAddedAsGDP':'d_WBD_SVAasGDP'
,'i_RuralPopulationAsTPopulation':'d_WBD_RPasTP'
,'i_GenderParityIndex':'d_WBD_GPI'
,'i_CO2EmissionsPerCapita':'d_WBD_co2pc'
,'i_Population Growth':'d_WBD_pg'
,'i_RenewableEnergyConsuption':'d_WBD_REC'
,'i_UnemploymentRate':'d_WBD_UR'
,'i_Expenditures on Education':'d_WBD_EonE'
,'i_TertiarySchoolEnrollmentAsPerce':'d_WBD_TSE %'
,'i_WomenLaborForceParticipation':'d_WBD_WLFP'
,'i_OreAndMetalsExports':'d_WBD_OME'
,'i_EnergyIMports':'d_WBD_energyI'
,'i_MarketCapitalizationAsGDP':'d_WBD_MCasGDP'
,'i_PublichHealthExpenditures':'d_WBD_PHE'
,'i_AgingPopulation':'d_WBD_AP'
,'i_MobilePhonesDiffusion':'d_WBD_MPD'
,'i_InternetServersperMillion':'d_WBD_ISpermil'
,'i_Air_transport_registered_carri':'d_WBD_ATR'
,'i_PortContainerTraffic':'d_WBD_PCT'
,'i_GINI':'d_WBD_GINI'
,'i_DisclosureIndex':'d_WBD_discloserI'
,'i_BanksAsSourceofCredit':'d_WBD_bankasSOC'
,'i_HighTechExports':'d_WBD_HTE'
,'i_InternationalTourismExpenditure':'d_WBD_ITE'
,'i_InternationalTourismReceipts':'d_WBD_ITR'
,'i_LeadTimeToExport':'d_WBD_leadtimeE'
,'i_MerchandiseTrade':'d_WBD_MT'
,'i_StrengthofLegalRights':'d_WBD_strengthLR'
,'i_NumberOfSTartupProcedures':'d_WBD_startup_procedures'
,'i_IntentionalHomicides':'d_WBD_IH'
,'i_NewBusinessFormation':'d_WBD_NBF'
,'i_ResearchAndDevelopemntSpending':'d_WBD_RandDspending'
,'i_PatentApplicationResidents':'d_WBD_patentapps'
,'i_TrademarkAppsResdients':'d_WBD_TMapps'
,'i_ResearchersinR_D':'d_WBD_researchersRD'
,'i_FuelExport':'d_WBD_FE'
,'i_Year_WBD':'d_WBD_year'
,'i_Year':'d_WBD_year1'})
    return df
    
```

```python
def change_columns(wbd_df,df_p):
    wbd_df = wbd_df.add_prefix('i_')
    wbd_df = wbd_df.rename(columns={'i_Country Name':'icoun'})
    df_p = df_p.merge(wbd_df, on='icoun',how='left')
    df_p = rename_icon_columns(df_p)
    wbd_df = wbd_df.rename(columns={'icoun':'dcoun'})
    df_p = df_p.merge(wbd_df, on='dcoun',how='left')
    df_p =  rename_dcon_columns(df_p)
    return wbd_df,df_p
```
