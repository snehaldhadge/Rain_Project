```python
import pandas as pd
import numpy as np
from datetime import datetime, date, time, timedelta
```

```python
def add_ipri_columns(df,ipri):
    ipri = ipri.rename(columns={'Countries':'dcoun'})
    df = df.merge(ipri, on='dcoun',how='left')
    df = df.rename(columns={'ipri':'ipri_d'})
    
    ipri = ipri.rename(columns={'dcoun':'icoun'})
    df = df.merge(ipri, on='icoun',how='left')
    return df
```

```python
df = pd.read_csv("final_projects_wbd_wgi.csv")
```

```python
ipri = pd.read_excel("International Property Rights Index.xlsx")
ipri
```

```python
ipri_cols =  [col for col in ipri if col.startswith('Score')]


ipri_selected = ipri[ipri_cols]

ipri_selected = ipri_selected.rename(columns = lambda x: x.split(' ')[1])
ipri_selected['Countries'] = ipri['Countries']
ipri_selected
```

```python
df_p2014 = df[df['adateyear'] == 2014]
df_p2015 = df[df['adateyear'] == 2015]
df_p2016 = df[df['adateyear'] == 2016]
df_p2017 = df[df['adateyear'] == 2017]
df_p2018 = df[df['adateyear'] == 2018]
df_p2019 = df[df['adateyear'] == 2019]
```

```python
ipri_2014 = ipri_selected[['Countries','2014']]
ipri_2014 = ipri_2014.rename(columns={'2014':'ipri'})
df_p2014 = add_ipri_columns(df_p2014,ipri_2014)
df_p2014.drop(['Unnamed: 0','Unnamed: 0.1'],axis=1,inplace=True)
```

```python
ipri_2015 = ipri_selected[['Countries','2015']]
ipri_2015 = ipri_2015.rename(columns={'2015':'ipri'})
df_p2015 = add_ipri_columns(df_p2015,ipri_2015)
df_p2015.drop(['Unnamed: 0','Unnamed: 0.1'],axis=1,inplace=True)
```

```python
ipri_2016 = ipri_selected[['Countries','2016']]
ipri_2016 = ipri_2016.rename(columns={'2016':'ipri'})
df_p2016 = add_ipri_columns(df_p2016,ipri_2016)
df_p2016.drop(['Unnamed: 0','Unnamed: 0.1'],axis=1,inplace=True)
```

```python
ipri_2017 = ipri_selected[['Countries','2017']]
ipri_2017 = ipri_2017.rename(columns={'2017':'ipri'})
df_p2017 = add_ipri_columns(df_p2017,ipri_2017)
df_p2017.drop(['Unnamed: 0','Unnamed: 0.1'],axis=1,inplace=True)
```

```python
ipri_2018 = ipri_selected[['Countries','2018']]
ipri_2018 = ipri_2018.rename(columns={'2018':'ipri'})
df_p2018= add_ipri_columns(df_p2018,ipri_2018)
df_p2018.drop(['Unnamed: 0','Unnamed: 0.1'],axis=1,inplace=True)
```

```python
ipri_2019 = ipri_selected[['Countries','2019']]
ipri_2019 = ipri_2019.rename(columns={'2019':'ipri'})
df_p2019 = add_ipri_columns(df_p2019,ipri_2019)
df_p2019.drop(['Unnamed: 0','Unnamed: 0.1'],axis=1,inplace=True)
```

```python
df_projects = df_p2014
df_projects = df_projects.append(df_p2015, ignore_index=True)
df_projects = df_projects.append(df_p2016, ignore_index=True)
df_projects = df_projects.append(df_p2017, ignore_index=True)
df_projects = df_projects.append(df_p2018, ignore_index=True)
df_projects = df_projects.append(df_p2019, ignore_index=True)
df_projects.to_csv('final_projects.csv')
```
