```python
import pandas as pd
import numpy as np
import random
import math
```

```python
def get_yearwise_proj_data(df,year):
    bvdid_list = df[df['adate_year']==year]['iBVDid']
    bvdid_list = bvdid_list.dropna()
    bvdid_list = bvdid_list.reset_index()
    bvdid_list.drop(['index'],axis=1,inplace=True)
    return bvdid_list
```

```python
def get_yearwise_rdate_proj_data(df,year):
    bvdid_list = df[df['rdate_year']==year]['iBVDid']
    bvdid_list = bvdid_list.dropna()
    bvdid_list = bvdid_list.reset_index()
    bvdid_list.drop(['index'],axis=1,inplace=True)
    return bvdid_list
```

```python
df2 = pd.read_csv("final_projects.csv")
```

# Get year wise BVDID

```python
df = df2
```

```python

df['adate_year'] = (pd.to_datetime(df['adate']).dt.year).astype('Int64')
```

```python
for year in df['adate_year'].unique():
    if not pd.isnull(year):
        bvdid_list=get_yearwise_proj_data(df,year)
        bvdid_list.to_csv(r'BDVID_List_'+str(year)+'.txt',header=None,index=None)
 
```

```python
r_df = df[df['adate_year'].isnull()]
```

```python
r_df['rdate_year'] = (pd.to_datetime(df['rdate']).dt.year).astype('Int64')
```

```python
r_df['rdate_year'].unique()
```

```python
for year in r_df['rdate_year'].unique():
    if not pd.isnull(year):
        bvdid_list=get_yearwise_rdate_proj_data(r_df,year)
        bvdid_list.to_csv(r'Rdate_BDVID_List_'+str(year)+'.txt',header=None,index=None)
```
