```python
def get_year_data(bmi_data,year):
    cols = [c for c in bmi_data.columns if c.startswith(year) or c.startswith('Unnamed')]
    new_df =  bmi_data[cols]
    new_df.columns= ['Country','both_sexes','male','female']
    new_df['adate_year'] = year
    return new_df[2:]
    
    
```

## Merge tcoun data using adate

```python
bmi_t = bmi_df
bmi_t.columns = ['tcoun','bmi_both_sexes_t','bmi_male_t','bmi_female_t','adateyear']
bmi_t
```

```python

bmi_t['adateyear']=bmi_t['adateyear'].astype(int)
deals['adateyear'] = deals['adateyear'].astype(int)
```

```python
a = pd.merge(deals, bmi_t,how="left" ,on=["tcoun","adateyear"]) 
a
```

```python
bmi_i = bmi_df
bmi_i.columns = ['icoun','bmi_both_sexes','bmi_male','bmi_female','adateyear']
bmi_i
```

```python
icoun_d = pd.merge(a, bmi_i,how="left" ,on=["icoun","adateyear"]) 
icoun_d
```
